```python
import cv2
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_prefix = 'Z://RR/Training_Info/FairFace_All_Races_Testing/loss'
save_path = 'Z://RR/Final/report_work/Training_Analysis_Images/'
```

```python
# fig = plt.figure(figsize=(22, 13))
fig, axs = plt.subplots(2, 3, figsize=(25, 12))
ax = axs.ravel()
plt.subplots_adjust(hspace=0.35,wspace=0.2)


for fold in range(6):
	
	file_name = F"results_ResNet34_v21_{fold}.csv"
	# path = os.path.join(file_prefix,file_name)
	path = file_prefix + '/' + file_name
	df =pd.read_csv(path)
 
	_ = sns.lineplot(data = df, x = 'epoch', y = 'val_loss',label = 'Validation Loss', ax = ax[fold])
	_ = sns.lineplot(data = df, x = 'epoch', y = 'train_loss',label = 'Training Loss', ax = ax[fold])
	ax[fold].set_xlabel('Epoch')
	ax[fold].set_ylabel('Loss')
	ax[fold].set_title(F"Loss Curve for Fold {fold}")
 
plt.savefig(os.path.join(save_path,'loss_curve.jpg'))
plt.show()
```

```python
# fig = plt.figure(figsize=(22, 13))
fig, axs = plt.subplots(2, 3, figsize=(25, 12))
ax = axs.ravel()
plt.subplots_adjust(hspace=0.35,wspace=0.2)


for fold in range(6):
	
	file_name = F"results_ResNet34_v21_{fold}.csv"
	# path = os.path.join(file_prefix,file_name)
	path = file_prefix + '/' + file_name
	df =pd.read_csv(path)
 
	_ = sns.lineplot(data = df, x = 'epoch', y = 'val_acc',label = 'Validation Accuracy', ax = ax[fold])
	_ = sns.lineplot(data = df, x = 'epoch', y = 'train_acc',label = 'Training Accuracy', ax = ax[fold])
	ax[fold].set_xlabel('Epoch')
	ax[fold].set_ylabel('Accuracy')
	ax[fold].set_title(F"Accuracy Curve for Fold {fold}")

plt.savefig(os.path.join(save_path,'acc_curve.jpg'))
plt.show()
```
